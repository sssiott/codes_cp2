```python
dict = {}
for id, brand in zip(cpy_1['product_id'],cpy_1['brand']):
    if id in dict:
        continue
    else:
        dict[id] = brand
```

```python
# 고유한 product_id와 brand명
# len(dict)
```

```python
# cp
```

```python
for i in range(len(cp)):
    if type(cp.loc[i,'brand']) != str:
        cp.loc[i,'brand'] = dict[cp.loc[i,'product_id']]
```

### 카테고리코드 결측치 & 컬럼 분할

```python
# dataset['category_code'].value_counts()
```

```python
cpy = dataset[['category_id','category_code']].copy()
```

```python
# cpy.isnull().sum()
```

```python
# cpy['category_id'].value_counts()
```

```python
# cpy.loc[cpy['category_id'] == 2053013555631882655]
```

```python
cpy_1 = cpy.sort_values('category_id').drop_duplicates()
```

```python
dict = {}
for id, code in zip(cpy_1['category_id'],cpy_1['category_code']):
    if id in dict:
        continue
    else:
        dict[id] = code
```

```python
# len(dict)
```

```python
for i in range(len(cpy)):
    if type(cpy.loc[i,'category_code']) != str:
        cpy.loc[i,'category_code'] = dict[cpy.loc[i,'category_id']]
```

```python
# cpy.isnull().sum()
```

* id에 연결된 code가 모두 null

```python
cpy.fillna('empty',inplace=True)
```

```python
# cpy
```

```python
# 카테고리 코드를 '.'을 기준으로 스플릿
li = []
for i in range(len(cpy)):
    li.append(cpy['category_code'].iloc[i].split('.'))
```

```python
# li
```

```python
# 1분류 2분류 3분류 각각 분할해서 리스트로 저장
cat_first = []
cat_second = []
cat_third = []
for i in li:
    cat_first.append(i[0])
    try:
        cat_second.append(i[1])
    except:
        cat_second.append(i[0])
        cat_third.append(i[0])
        continue
    try:
        cat_third.append(i[2])
    except:
        cat_third.append(i[1])
```

```python
dataset['first_level'] = cat_first
dataset['secont_level'] = cat_second
dataset['third_level'] = cat_third
```

```python
# 기존 event_time과 category_code 컬럼 삭제
dataset.drop(['event_time','category_code'],axis=1,inplace=True)
```
