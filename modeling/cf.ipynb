```python
import pandas as pd
from tqdm import tqdm
url = '../dataset/'
df = pd.read_csv(url+'dataset_light.csv')
```

```python
df.head()
```

```python
cpy = df[['user_id','product_id','event_type']].groupby(['user_id','product_id','event_type']).size().reset_index()
```

```python
cpy[cpy['event_type']== 'purchase']
```

```python
score_list = []
for user in tqdm(cpy['user_id'].drop_duplicates()):
    user_1 = cpy[cpy['user_id'] == user]
    for product in user_1['product_id'].drop_duplicates():
        product_1 = user_1[user_1['product_id'] == product]
        
        if 'purchase' in product_1['event_type'].to_list():
            # if product_1[product_1['event_type'] == 'purchase'].iloc[:,-1] > 2:
            #     score_list.append([user,product,4])
            # else:
            #     score_list.append([user,product,3])
            score_list.append([user,product,3])
        elif 'cart' in product_1['event_type'].to_list():
            score_list.append([user,product,2])
        else:
            score_list.append([user,product,1])
```

```python
rate_df = pd.DataFrame(score_list,columns=['user_id','product_id','rate'])
```

```python
rate_df.to_csv(url+'rate.csv',index=False)
```
